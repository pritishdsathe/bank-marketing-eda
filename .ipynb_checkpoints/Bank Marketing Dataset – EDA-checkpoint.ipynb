## üè¶ Overall Summary: Loan Types and Subscription Behavior

Understanding customer loan profiles gives deep insight into financial status and their likelihood of subscribing to bank services.

### 1. People with Both Housing and Personal Loan
There are 406 customers who hold both housing and personal loans. This group typically represents individuals with higher financial commitments.

### 2. People with Only Housing Loan
A large segment of customers have only housing loans, indicating long-term asset investments like property or real estate.

### 3. People with Only Personal Loan
Another segment consists of those with only personal loans. These loans may represent short-term financial needs or consumer spending.

### 4. People with No Loans
A considerable number of customers do not hold any loans. This group potentially represents financially stable individuals or those less reliant on credit.

### 5. Subscription Distribution Compared with Loan Groups
When comparing subscription behavior across these loan categories, it was observed that customers **without any loans** had a **higher subscription rate**. This may be due to reduced financial burden and more willingness to engage with new offers.

### 6. Subscription Rate Comparison
A pie chart showcasing the distribution among subscribers revealed that:
- The **highest** subscription share came from customers with **no loans**.
- The **lowest** subscription share came from customers who had **both housing and personal loans**.


```python
len(bank_record[(bank_record['housing'] == 'yes') & (bank_record['loan'] == 'yes')])
```

```python
len(bank_record[(bank_record['housing'] == 'no') & (bank_record['loan'] == 'no')])
```

```python
bank_record["Loan Type"] = "No Loans"
bank_record.loc[(bank_record['housing'] == 'yes') & (bank_record['loan'] == 'no'), 'Loan Type'] = 'Housing Loan Only'
bank_record.loc[(bank_record['housing'] == 'no') & (bank_record['loan'] == 'yes'), 'Loan Type'] = "Personal Loan Only"
bank_record.loc[(bank_record['housing'] == 'yes') & (bank_record['loan'] == 'yes'), 'Loan Type'] = 'Both Loans'

plt.figure(figsize=(10,6))
sns.countplot(data=bank_record, x='Loan Type', hue='y')
plt.title("Subscription status by loan group")
plt.tight_layout()
plt.savefig('Plots/Subscription_status_by_loan_group')
plt.show()
```

```python
# Filtered only those who subscribed
subscribed = bank_record[bank_record['y'] == 'yes']

# Count of each loan type among subscribers
loan_counts = subscribed['Loan Type'].value_counts()

# Plot
plt.figure(figsize=(10,6))
plt.pie(loan_counts, labels=loan_counts.index, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightgreen', 'orange', 'pink'])
plt.title("Subscription Distribution by Loan Type")
plt.axis('equal')  # Circle shape
plt.savefig('plots/subscription_pie_by_loan_type.png')
plt.show()
```

```python
subset = bank_record[(bank_record['housing'] == 'no') & (bank_record['loan'] == 'no')].copy()
subset['y'].value_counts(normalize=True)
```

```python
subset2 = bank_record[(bank_record['housing'] == 'yes') & (bank_record['loan'] == 'yes')].copy()
subset2['y'].value_counts(normalize=True)  # Subscription rate in this group
```

```python
# Compare subscription rate
group_all_loans = subset2['y'].value_counts(normalize=True)
group_no_loans = subset['y'].value_counts(normalize=True)
```

```python
group_all_loans
```

```python
group_no_loans
```

```python
subs_rate = pd.DataFrame({'Both Loans': group_all_loans, 'No Loans': group_no_loans}).T
subs_rate.plot(kind='bar', stacked=False)
plt.ylim(0,1)
plt.title("Subscription Rate Comparision")
plt.ylabel("Proportion")
plt.savefig('Plots/Subscription Rate Comparision.png')
plt.show()
```

### ‚úÖ Conclusion

Customers with **fewer financial obligations** tend to be more responsive to marketing campaigns. Subscription rates were highest among those **without any loans** or **with only one loan type**, highlighting the importance of financial status when targeting leads. Banks can utilize this insight to refine their outreach strategies and improve conversion efficiency.

# üìä Monthly Campaign Summary: Contacts, Subscriptions & Rates
## 1. Contacts Made per Month
The number of contacts made to potential customers varied month to month. Peak months like May and July saw the highest outreach, reflecting focused campaign efforts during these periods.

## 2. Number of People Subscribed per Month
The number of customers who subscribed after contact also changed monthly. While more contacts generally meant more subscriptions, this raw count doesn‚Äôt fully capture how effective the campaigns were.

## 3. Subscription Rate per Month
The subscription rate ‚Äî the percentage of contacted people who subscribed ‚Äî reveals the campaign‚Äôs true efficiency. Some months with fewer contacts achieved higher subscription rates, indicating better targeting or messaging success.

```python
bank_record.groupby('month').size().sort_values(ascending=False)
```

```python
bank_record.groupby(['month', 'y']).size().sort_values(ascending=False).reset_index(name='Count of People')
```

```python
subscription_rate_per_month = bank_record.groupby('month')['y'].value_counts(normalize=True).unstack()['yes'] * 100
subscription_rate_per_month = subscription_rate_per_month.reset_index(name='Subscription Rate (%)')
```

```python
subscription_rate_per_month.sort_values(by='Subscription Rate (%)', ascending=False)
```

```python
plt.figure(figsize=(10,4))
plt.plot(subscription_rate_per_month['month'], subscription_rate_per_month['Subscription Rate (%)'], marker='o', linestyle='-', color='red')
plt.title("Monthly Subscription Rate")
plt.xlabel("Months")
plt.ylabel("Subscription Rate (%)")
plt.tight_layout()
plt.savefig('Plots/Monthly_Subscription_Rate.png')
plt.show()
```

# üéØ Conclusion and Key Insights  

Peak contact months (May, July) produced the greatest number of subscriptions, showing the importance of timing in campaign outreach.

Higher subscription rates in December, March and October suggest more effective campaigns during these months, possibly due to better customer engagement or offers.

# üìû Summary of Contact Methods Used in Bank Campaigns

## 1. Count of Different Contact Methods

The bank contacted customers primarily through three methods:

 - *Cellular (mobile calls)*

 - *Telephone (landline calls)*

 - *Unknown (cases where cont*act method was not recorded)

Analysis shows that cellular calls were the most frequently used method, accounting for the majority of customer contacts.

# 2. Contact Method Generating the Highest Number of Leads
When examining lead generation (customers who subscribed), cellular calls also generated the highest number of positive responses. This indicates that mobile outreach was not only the most common but also the most effective contact method.


# üîë Key Insights
- Cellular calls dominate both in frequency and lead generation, making them the most impactful channel for the bank.

- Telephone calls have lower usage and fewer leads in comparison.

- The unknown category mostly represents missing or unrecorded contact methods, which should be minimized for better data accuracy.

```python
bank_record['contact'].value_counts()
```

```python
highest_lead_generation = bank_record[bank_record['y'] == 'yes']['contact'].value_counts()
```

```python
highest_lead_generation_df = highest_lead_generation.reset_index()
highest_lead_generation_df.columns = ['contact_type', 'lead_count']  # rename columns
```

```python
highest_lead_generation_df
```

# üéØ Recommendations and Conclusion

- Continue prioritizing and possibly expanding cellular contact efforts to maximize campaign effectiveness.

- Explore ways to improve telephone contact strategies to boost their contribution.

- Work on reducing the ‚Äòunknown‚Äô contact records by improving data quality and tracking methods.

**Conclusion: Focusing on cellular outreach has clearly been beneficial for the bank‚Äôs marketing efforts. The data shows a strong correlation between mobile contacts and lead generation success, emphasizing the importance of this channel.**

### ‚úÖ Key Takeaways:

   * The number of contacts made in each campaign had a **negative correlation** with subscription. Excessive contacts may reduce interest.
   * **May** had the **highest contact volume**, but also had one of the **lowest subscription rates**, indicating poor timing or ineffective communication during that month.

3. **Communication Channels:**

   * Among different contact methods, **cellular calls led to the most subscriptions**.
   * Campaigns using **telephone or unknown contact types** had lower effectiveness.

4. **Job-Based Patterns:**

   * **Blue-collar** workers were the largest group but had the **lowest subscription rates**.
   * **Management and admin.** showed moderate subscription potential.
   * Loan ownership varied by job‚Äî**blue-collar and services** workers held the most housing loans, while **retired and students** mostly did not.

5. **Education Impact:**

   * People with **tertiary education** had the **highest subscription rate (14.3%)**, even though **secondary education** was the most common.
   * A clear positive trend was observed between education level and likelihood of subscription.

6. **Financial Factors:**

   * **366 customers had negative balances**, with the **majority being blue-collar and management**.
   * **91.5% of these customers did not subscribe**, showing that financial stress may strongly affect campaign conversion.

7. **Loan Ownership & Subscription:**

   * Customers with **no loans** had slightly better subscription rates compared to those with **both housing and personal loans**.
   * This may suggest that existing loan commitments reduce the willingness to take on new financial products.
