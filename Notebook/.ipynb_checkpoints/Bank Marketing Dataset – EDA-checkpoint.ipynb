```python
# Filtered only those who subscribed
subscribed = bank_record[bank_record['y'] == 'yes']

# Count of each loan type among subscribers
loan_counts = subscribed['Loan Type'].value_counts()

# Plot
plt.figure(figsize=(10,6))
plt.pie(loan_counts, labels=loan_counts.index, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightgreen', 'orange', 'pink'])
plt.title("Subscription Distribution by Loan Type")
plt.axis('equal')  # Circle shape
plt.savefig('plots/subscription_pie_by_loan_type.png')
plt.show()
```

```python
subset = bank_record[(bank_record['housing'] == 'no') & (bank_record['loan'] == 'no')].copy()
subset['y'].value_counts(normalize=True)
```

```python
subset2 = bank_record[(bank_record['housing'] == 'yes') & (bank_record['loan'] == 'yes')].copy()
subset2['y'].value_counts(normalize=True)  # Subscription rate in this group
```

```python
# Compare subscription rate
group_all_loans = subset2['y'].value_counts(normalize=True)
group_no_loans = subset['y'].value_counts(normalize=True)
```

```python
group_all_loans
```

```python
group_no_loans
```

```python
subs_rate = pd.DataFrame({'Both Loans': group_all_loans, 'No Loans': group_no_loans}).T
subs_rate.plot(kind='bar', stacked=False)
plt.ylim(0,1)
plt.title("Subscription Rate Comparision")
plt.ylabel("Proportion")
plt.savefig('Plots/Subscription Rate Comparision.png')
plt.show()
```

# üí° Insights

**‚úÖ No Loans = Higher Subscription Ratio**
The group without loans shows the highest ratio of yes responses **(16.9%)**, indicating stronger interest or financial ability to commit.

üè† + üí≥ Both Loans = Lowest Conversion
Clients holding both loans have the lowest subscription ratio **(6.2%)**, suggesting financial obligations may deter further commitments.

üßÆ Skewed Subscription Distribution
Over **54%** of all subscribers had **no loans**. Clients with **only housing loans** contributed **37.4%**, while those with any **personal loan (alone or with housing) made up just 8.3% combined.**

3. ‚òéÔ∏è Telephone Has Limited Reach and Impact
Telephone contacts are both low in volume and less effective in lead generation, suggesting possible deprioritization or reconsideration.

### Conclusion
* Prioritize cellular channels for outreach campaigns to maximize lead generation.

This comprehensive exploratory data analysis on the bank marketing dataset uncovers critical insights that inform strategic decision-making:

- **Customer Segmentation & Behavior:**

- **Data Quality & Opportunities:**  
  Segments marked as ‚Äòunknown‚Äô in job and education categories represent an area for data enrichment, as they surprisingly show moderate to high subscription rates.
